```python
import pandas as pd

data_a = pd.DataFrame({
    "age": [20, 25, 30, 35, 40],
    "salary": [20000, 30000, 40000, 50000, 60000],
    "target": [1, 0, 1, 0, 1]
})

data_b = pd.DataFrame({
    "experience": [1, 2, 4, 6, 8],
    "education": [10, 12, 14, 16, 18],
    "target": [1, 0, 1, 0, 1]
})
```

```python
from goseidataset import Supervised_learning

sl = Supervised_learning(
    data_a,
    data_b,
    target="target"
)
```

```python
merged = sl.merge_same_target()

print(merged)
```

```python
from sklearn.ensemble import RandomForestRegressor

result = sl.compare_models(
    model=RandomForestRegressor()
)

result
```

```python
result = sl.weighted_ensemble(
    model=RandomForestRegressor()
)

result.keys()
```

```python
data_a.loc[1, "salary"] = None
data_b.loc[2, "education"] = None

sl = Supervised_learning(
    data_a,
    data_b,
    target="target"
)
```

```python
result = sl.vertical_stack(
    method="mean"
)

result["dataset"]
```

```python
result["dataset"].isnull().sum()
```

```python
result = sl.vertical_stack(
    method="median"
)

result["dataset"]
```

```python
result = sl.vertical_stack(
    method="mode"
)

result["dataset"]
```

```python
result = sl.vertical_stack(
    method="knn",
    n_neighbour=3
)

result["dataset"]
```

```python
result = sl.vertical_stack(
    method="iterative"
)

result["dataset"]
```

```python
import numpy as np

data_a = pd.DataFrame({
    "age": np.random.randint(20,60,100),
    "experience": np.random.randint(1,20,100),
    "salary": np.random.randint(30000,100000,100)
})

data_a.loc[0:20,"salary"] = np.nan

data_b = data_a.copy()

sl = Supervised_learning(
    data_a,
    data_b,
    target="salary"
)
```

```python
result = sl.vertical_stack(
    method="random_forest",
    predictors={
        "salary": [
            "age",
            "experience"
        ]
    }
)

result["report"]
```

```python
result = sl.vertical_stack(
    method="xgboost",
    predictors={
        "salary": [
            "age",
            "experience"
        ]
    }
)

result["report"]
```
